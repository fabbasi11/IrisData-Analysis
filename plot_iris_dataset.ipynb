
# The Iris Dataset

```python
import pandas as pd

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])

# Adding target and species columns
df['target'] = iris.target
df['species'] = iris.target_names[iris.target]

# Descriptive analysis for 'petal length (cm)' and 'petal width (cm)'
petal_df = df[['petal length (cm)', 'petal width (cm)']]
petal_desc_stats = petal_df.describe()

# Print descriptive statistics
print("Descriptive Statistics for Petal Length and Petal Width:")
print(petal_desc_stats)

import matplotlib.pyplot as plt

# Histograms for Petal Length and Petal Width
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(df['petal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')

plt.subplot(1, 2, 2)
plt.hist(df['petal width (cm)'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = petal_df.corr()

# Print correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

import seaborn as sns

# Correlation heatmap for Petal Length and Petal Width
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix - Petal Length & Petal Width')
plt.show()
```

```python
from scipy.stats import ttest_ind

# Separate petal length for different species
setosa_petal_length = df[df['species'] == 'setosa']['petal length (cm)']
versicolor_petal_length = df[df['species'] == 'versicolor']['petal length (cm)']

# Perform t-test
t_statistic, p_value = ttest_ind(setosa_petal_length, versicolor_petal_length)

# Print t-test results
print("\nT-test for Difference in Petal Length between Setosa and Versicolor:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Boxplot for Petal Length by Species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('Boxplot of Petal Length by Species')
plt.show()
```
